# **Marco teórico**
El rendimiento académico es un aspecto que depende la interacción de factores personales, familiares, escolares y sociales. Desde la psicología educativa y las ciencias del aprendizaje se muestra que el desempeño de un estudiante no solo se puede explicar por sus capacidades como individuo, sino que también por el ambiente o contexto en el que se desarrolla. Variables como el apoyo familiar, acceso a recursos, motivación, calidad del profesorado y la interacción con otros compañeros influyen de forma directa o indirecta en los resultados escolares entonces las variables categóricas y cualitativas del entorno del estudiante son indicadores muy importantes para entender por qué varia su rendimiento.

Investigaciones anteriores como los estudios PISA y los trabajos de Bloom realizados en 1964 han concluido en que factores como la participación de los padres, el ingreso de la familia y el nivel educativo de los padres puede influir en el desempeño escolar ya sea porque facilitan recursos académicos, promueven hábitos de estudio o hacen que haya un ambiente de apoyo y de igual forma factores como la motivación, influencia de los compañeros y la calidad que tienen los profesores se relacionan con la actitud del estudiante hacia el aprendizaje, el que tanta dedicación le da al estudio y la persistencia ante dificultades, en otros estudios igual se señala que variables estructurales como el tipo de escuela, el acceso a internet o la distancia entre la escuela y la casa pueden dificultar la oportunidad de estudiar.

Para poder analizar estas variables es importante el uso de técnicas estadísticas que ayuden a comparar medias de distintos grupos, pero para esto es necesario conocer que variables se estaran estudiando.


**Rendimiento educativo (puntaje_examen - variable dependiente):** Se refiere no nadamas a los resultados cuantificables como calificaciones o puntajes de examen, sino también al logro de competencias y habilidades cognitivas, sociales y emocionales esperadas para un nivel educativo.


*   Puntajes estandarizados de pruebas de conocimientos
*   Promedios de calificaciones insititucionales

*   Variables binarias como aprobado o no aprobado.

**Factores asociados al rendimiento (Variables independientes):** Agrupados en niveles


*   **Factores individuales (Características propias al estudiante o bajo su control directo):**  
hora_estudio, horas_sueño, calificaciones_previas, nivel_motivación, influencia_compañeros, actividad_fisica, discapacidad_aprendizaje, genero.
*   **Factores escolares (Características de la institución educativa y su funcionamiento):** asistencia, acceso_a_recursos, actividades_extracurriculares, sesiones_tutoria, calidad_profesorado, tipo_escuela.


*   **Factores contextuales o familiares (Carterísticas del entorno externo del estudiante y la escuela):** participacion_parental, ingreso_familiar, acceso_internet, nivel_estudio_padres, distancia.

**Pruebas estadísticas**

**ANOVA:**

El ANOVA es una herramienta que se usa para determinar si existen diferencias significativas entre los niveles de una variable categórica con respecto a una variable cuantitativa en ese caso sería el puntaje del examen, da respuesta a la hipótesis nula del ANOVA que dice que todas las medias de los grupos son iguales y cuando el resultado es significativo se puede concluir que al menos un grupo es diferente a otros.

Pero ANOVA no puede identificar por sí mismo entre que grupos se encuentran esas diferencias por eso cuando se llega a observar que hay significancia se usa la prueba Tukey.

**Tukey:**

La prueba Tukey permite comparar cada par de categorías de manera más precisa y controla el error por comparaciones múltiples, es adecuada en estudios que incluyen varios niveles categóricos porque nos dice cuales grupos muestran un mejor o peor desempeño académico.

**Eta cuadrada ($\eta^2$):**

Un elemento importante del análisis es el tamaño del efecto el cual es representado por el estadístico eta cuadrada $\eta^2$ el cual dice la proporción de la variabilidad total del rendimiento que es explicada por una variable categórica, aunque un factor puede ser significativo el tamaño del efecto puede ser pequeño esto refleja que su influencia sobre el puntaje es limitada. Esto puede ser importante dentro de lo educativo porque de forma frecuente los efectos de variables externas son más moderados debido a que son diversos factores los que influyen.


# **Metodología**

En el presente estudio se desarrollara en base a una base de datos que contiene información de miles de estudiantes, incluyendo su porcentaje de examen y distintas variables categóricas relacionadas con características personales, familiares y escolares. Antes de el desarrollo de algún procedimiento estadístico se eliminara del vaciado de datos, registros que llegaran a faltar para garantizar que todas las variables categóricas esten bien codificadas y con niveles consistentes como puede se bajo, medio, elevado o en respuestas binarias como si y no.

Ya lista la base de datos, el análisis se basara solamente en al relación de variables categóricas y el rendimiento académico que sera medido con el puntaje del examen. Para poder analizar estas relaciones se utilizara ANOVA la cual permite identificar si hay diferencias significativas entre las medias de los diferentes niveles de cada variable categórica, es buena opción cuando se analizan varios factores al mismo tiempo por que evalúa el efecto individual de cada variable aunque estén presentes otras.

Cuando ANOVA arroje que hay diferencias significativas en una variable, se aplicara la prueba Tukey esto para comparar cada par de categorías, ya que sera importante aplicarla para determinar cuales niveles de las variables tendran diferencias con el puntaje del examen y así ver que grupos tendran mejor o peor desempeño académico, su uso es buena opción cuando se realizan estudios con variables categóricas por que tiene un control sobre el error cuando se hacen varias comparaciones y da intervalos de confianza para cada diferencia de medias.

Al final para poder completar el estudio de los resultados obtenidos y poder evaluar que tan relevante es la practica de cada variable se calculara el tamaño del efecto con el estadístico es cuadrado $\eta^2$, permitira identificar que tanta variabilidad total del puntaje del examen es explicada por cada variable haciendo diferencia entre variables con significancia pero con un impacto bajo y las que tegan un impacto limitado demostraran una influencia más notable en el rendimiento académico, con todos estos procedimientos se podra analizar la influencia de factores categóricos y establecer conclusiones.

## **Vaciado de datos**

```python
 import pandas as pd
 url = "https://raw.githubusercontent.com/ag0466154-cpu/Unidad-4_Estadistica2025B-/refs/heads/main/dataset.csv"
 df = pd.read_csv(url)
 df = df.dropna()
 df
```

## **Tabla ANOVA**

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm

# Nivel de significancia = 0.05
# H0: Todas las medias son iguales
# H1: Al menos una media es distinta

# Realizar el modelo ANOVA de dos vías
modelo_lineal = ols(f'puntaje_examen ~ 	horas_estudio + asistencia + C(participacion_parental) + C(acceso_a_recursos) + C(actividades_extracurriculares) + horas_suenio + calificaciones_previas + C(nivel_motivacion) + C(acceso_internet) + sesiones_tutoria + C(ingreso_familiar) + C(calidad_profesorado) + C(tipo_escuela) + C(influencia_companieros) + actividad_fisica + C(	discapacidad_aprendizaje) + C(nivel_estudio_padres) + C(distancia) + C(genero)', data=df).fit()
tabla_anova = sm.stats.anova_lm(modelo_lineal, typ=2)
tabla_anova
```

## **Participación parental**

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df["participacion_parental"], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Participación parental", xlabel="Puntaje del examen")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
```

```python
# η² (eta cuadrado)	Tamaño del efecto
# ~0.01	              Pequeño
# ~0.06	              Mediano
# ~0.14 o más	        Grande

# Se interpreta como el coeficiente de determinación

# Estos valores pueden variar según la disciplina. En estudios de ciencias
# sociales, un η² de 0.06 puede ser muy significativo;
# en física o ingeniería, quizá se espera más.

eta_squared = tabla_anova['sum_sq']['C(participacion_parental)'] / tabla_anova['sum_sq'].sum()
print(f"η² (tamaño de efecto): {eta_squared:.3f}")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df["genero"], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Genero", xlabel="Puntaje de examen")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
```
